```python
penguins_full = penguins.dropna()

y = penguins_full.body_mass_g

features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm']
X = penguins_full[features]
```

```python
penguin_model = DecisionTreeRegressor(max_depth = 2, random_state=42)

model_fit = penguin_model.fit(X, y)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(model_fit, out_file='penguins_dt.dot',feature_names=features,rounded=True,filled=True)
Source.from_file('penguins_dt.dot')
```

```python
penguin_model = DecisionTreeRegressor(max_depth = 4, random_state=42)

model_fit = penguin_model.fit(X, y)
export_graphviz(model_fit, out_file='penguins_dt.dot',feature_names=features,rounded=True,filled=True)
Source.from_file('penguins_dt.dot')
```

```python
penguin_model = DecisionTreeRegressor(random_state=42)

model_fit = penguin_model.fit(X, y)
from sklearn.tree import export_graphviz

export_graphviz(model_fit, out_file='penguins_dt.dot',feature_names=features,rounded=True,filled=True)
Source.from_file('penguins_dt.dot')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state = 985) 
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
cart = DecisionTreeClassifier()
cart_tuned = DecisionTreeClassifier(max_depth = 10, min_samples_split = 54).fit(X_train, y_train)
y_pred = cart_tuned.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
cart_tuned = DecisionTreeClassifier(max_depth = 20, min_samples_split = 54).fit(X_train, y_train)
y_pred = cart_tuned.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
